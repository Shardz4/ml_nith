```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class RidgeRegression:
    def __init__(self, alpha=1.0):
        self.alpha = alpha
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        n_features = X.shape[1]
        X_b = np.c_[np.ones((len(X), 1)), X]
        I = np.eye(n_features + 1)
        I[0, 0] = 0 # Don't penalize the bias
        
        theta = np.linalg.inv(X_b.T.dot(X_b) + self.alpha * I).dot(X_b.T).dot(y)
        self.bias = theta[0]
        self.weights = theta[1:]
        
    def predict(self, X):
        return X.dot(self.weights) + self.bias
```

```python
if __name__ == "__main__":
    # 1. Load Data
    df = pd.read_csv('laptop_prices.csv')
    csv_columns = ['cpu_cores', 'ram_gb', 'gpu_memory_gb', 'sticker_count']
    X_data = df[csv_columns].values 
    y_data = df['price_usd'].values
    feature_names = ['CPU', 'RAM', 'GPU', 'Stickers']

    # 2. Train Model
    model = RidgeRegression(alpha=50.0)
    model.fit(X_data, y_data)
    
    # 3. Make Predictions
    predictions = model.predict(X_data)
```

```python
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # Plot A: The Weights
    ax1.bar(feature_names, model.weights, color='#1f77b4')
    ax1.axhline(0, color='black', linewidth=1)
    ax1.set_title('Ridge (L2) - Learned Weights')
    ax1.set_ylabel('Weight Value')

    # Plot B: Actual Regression (Actual vs Predicted)
    ax2.scatter(y_data, predictions, color='blue', label='Predictions')
    # Draw the perfect 45-degree prediction line
    min_val, max_val = min(y_data), max(y_data)
    ax2.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Fit')
    ax2.set_title('Ridge - Actual vs Predicted Prices')
    ax2.set_xlabel('Actual Price (USD)')
    ax2.set_ylabel('Predicted Price (USD)')
    ax2.legend()

    plt.tight_layout()
    plt.show()
```
